## Research Question

Which music genre exhibits the strongest positive and negative associations with a person’s mental health profile, specifically across self-reported Anxiety, Depression, and Insomnia scores on a 0-10 scale? 

This project makes use of statistical inference to determine if identifying a specific genre as one’s favorite and its average bpm correlates with significantly higher or lower mental health scores. By controlling age and hours of daily listening, we will isolate the relationship between musical preference and mental health to ensure that we are strictly analyzing the correlation between the two. This analysis aims to uncover whether certain genre preferences can be used as indicators of psychological distress patterns.  

## Background and Prior Work

Mental health among college students has become an increasingly urgent public health concern, anxiety, depression, and chronic stress rates being reported in recent years. In a 2024 report by the American College Health Association, 60% of college students reported overwhelming anxiety while another portion reported being functionally limited due to depression [1]. The research topic is necessary to see how college students utilize music as a low-barrier mental health service to cope with the daily stressors of life. Many college students do not seek formal mental health resources due to monetary constraints and the stigma around these struggles, and music has emerged as a unique outlet for emotional regulation. Listening to music has transformed from passive listening to an active tool for stress management and crisis aversion. Furthermore, identifying the relationship between music genres and mental health can provide insights into how different musical content can either contribute to or detract from mental health struggles when used as a tool for self-remedy. In recent years, studies have shown that music is beneficial to students with high stress levels. For example, a 2025 study titled “Statistical Insights into the Influence of Music on Mental Health” highlights the impact of musical engagement on psychological well-being by giving quantitative evidence [2].


Unlike therapy or medication, music requires no appointment, no cost, and no disclosure. Music is a low-barrier outlet for students, as it is already embedded in their daily life. The use of music as an emotional regulator serves as a prompt for researchers to look beyond music as entertainment but rather a legitimate coping mechanism. A thesis from the University of Guelph by Parya Abadeh argues that while music is a regular tool for emotional processing, the type of engagement determines whether the outcome is beneficial or not. Abadeh explains that “certain songs-especially those with negative or ambiguous emotional content- can reinforce rumination and other maladaptive emotion-regulation patterns, ultimately elevating stress…” [3]. While this study affirms that music listening broadly correlates with positive mental health outcomes, it treats musical engagement as a relatively uniform behavior, leaving open the question of whether what students listen to matters as much as the fact that they are listening at all.


However, the correlation between them is debated profusely, especially regarding certain genres of music and their levels of emotional intensity and calmness. A 2022 study by Kuo Wang, Sunyu Gao, and Jianhao Huang titled “Learning About Your Mental Health From Your Playlist? Investigating the Correlation Between Music Preference and Mental Health of College Students” found that preference for the pop music genre indicated less reported mental health struggles while heavy music preference showed an inverse relationship. Their results indicated the “correlation was significant and positive between pop music and mental health.” [4] They also observed a “significant and inverse relationship between college students’ preference for heavy music and their mental health”. [4] These findings are compelling, but leave some questions unanswered, since it remains unclear whether genre preference has a causative or consequential relationship to psychological stress patterns. Additionally, it is also unclear whether listening frequency of certain genres can play a role in regulating or destabilizing mental health struggles. With these posed questions, this project aims to address these gaps by holistically examining the relationship between student listening habits and mental health struggles to see how music can be utilized as a tool or deterrent for these emotional hardships.

## Hypothesis


We hypothesize that high BPM genres such as Rock or Metal will show a significant positive correlation with ‘high arousal’ symptoms, specifically Anxiety and Insomnia scores. We further predict that Depression scores will be high across both extremes of tempo in genres reflecting people’s different coping strategies. 

Fast auditory stimuli can increase heart rate and can potentially cause restlessness, making it harder to fall asleep. For this reason, we believe that high BPM music would cause more anxiety and insomnia. Additionally, people usually listen to music that reflects their current mood. Therefore, someone with Depression might listen to slow music to reflect their sadness, or they might pick faster, louder music as a way to let out frustration.

### Dataset: Music & Mental Health Results
1. The dataset “Music & Mental Health Survey Results” main independent variable is “Fav genre”, a variable that represents the respondent’s favorite music style, such as Rock, Pop, Metal, etc. To measure daily listening time for music, we use “Hours per day” to quantify in hours. Our dependent variables are “Anxiety”, “Depression”, and “Insomnia”. They are self-reported scores measured on a scale from 0 to 10. 0 indicates no symptoms of said variable, while 10 indicates extreme symptoms. Although these are highly subjective, higher scores indicate significant distress. We are also utilizing age in years as a control variable to find generational trends that could skew our results. We are also utilizing “BPM” to test if faster tempos correlate with high arousal. To specify, there are 736 observations and 33 variables.
2. There are several concerns about this dataset. The data does not have a large sample size (n = 736), especially for the older population, due to the distribution via online forums. There are genres that could have too few observations for us to figure out significant results. The self-reported scores are highly subjective because they are not clinically assessed. For example, one respondent’s specific score could be another’s lower score.



```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Loading Data
raw_path = 'data/00-raw/mxmh_survey_results.csv'
df = pd.read_csv(raw_path)

# Verifying tidyness: 
# This dataset is tidy because each row is a single respondent and each column is a single variable.
# We are also checking for potential duplicate entries.
# We verify this by checking for duplicate entries.
print(f"Number of duplicate rows: {df.duplicated().sum()}")
display(df.head())

# Data selection
# We are keeping these 7 columns as they are the only variables needed to satisfy our hypothesis. 
relevant_cols = ['Age', 'Hours per day', 'BPM', 'Fav genre', 'Anxiety', 'Depression', 'Insomnia']
df = df[relevant_cols]
```

```python
# Missingness Analysis (Pre-Cleaning)

# Rubric Requirement: Demonstrate where data is missing and analyze relationships
# The original shape of the data set
print(f"Initial Dataset Shape: {df.shape}")

# Showing what and where the data is missing while also showing potential relationships.
# Just from looking, it seems like many people did not know what the BPM of their favorite genre.
plt.figure(figsize=(10, 4))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Initial Missing Data Heatmap (Yellow = Missing)')
plt.show()
```

```python
# Checking the original number of entries and the number of entries missing for BPM 
initial_count = len(df)
bpm_missing = df['BPM'].isnull().sum()

# Drop rows missing critical labels
# We drop rows missing BPM or MH scores because these are important to 
# our hypothesis. Allowing these values to go through would risk creating false correlations.
df_clean = df.dropna(subset=['BPM', 'Anxiety', 'Depression', 'Insomnia', 'Fav genre']).copy()

# Remove outliers and potential troll entries
# The reason we filter is that, since these are survey results, there is
# no guarantee that people would not joke around and put random answers.

plt.figure(figsize=(12, 4))

# Check BPM for suspicious entries 
plt.subplot(1, 2, 1)
sns.boxplot(x=df['BPM'], color='skyblue')
plt.title('BPM Distribution Check')

# Check Age for suspicious entries 
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Age'], color='salmon')
plt.title('Age Distribution Check')

# Filtering BPM to a realistic range (40-250) and Age to < 95
# ensures our statistical inference isn't skewed by improbable data points.

df_clean = df_clean[(df_clean['BPM'] >= 40) & (df_clean['BPM'] <= 250)]
df_clean = df_clean[df_clean['Age'] < 95]

# Verifying the cleanliness of the dataset (Post-Cleaning) 
plt.figure(figsize=(10, 2))
sns.heatmap(df_clean.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Verification: Data has no missing values (Zero Nulls)')
plt.show()
```

```python
# Summary and Results
print(f"Original rows: {initial_count}")
print(f"Rows with missing BPM: {bpm_missing}")
print(f"Final usable rows after all cleaning: {len(df_clean)}")

print("\n--- Summary Statistics for Key Variables ---")
display(df_clean.describe())
```

```python
# Save Processed Data 
os.makedirs('data/02-processed', exist_ok=True)
df_clean.to_csv('data/02-processed/mxmh_cleaned.csv', index=False)
print("\nWrangled data saved to: data/02-processed/mxmh_cleaned.csv")
```
